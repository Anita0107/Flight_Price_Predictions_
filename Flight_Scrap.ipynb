```python
# 1.

price_info = []
route_info = []
airline=[]
journey=[]
source=[]
destination=[]
dep_time=[]
arrival_time=[]
duration=[]
total_stops=[]
flight_no=[]

# Airline
try:
    rom = driver.find_elements_by_xpath("//span[@class='txt-r4 ng-binding']")        
    for i in rom:
        airlines=i.text
        airline.append(airlines)
except NoSuchElementException:
    airline.append('-')
    
#source:
try:
    rom = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[2]")        
    for i in rom:
        sources=i.text
        source.append(sources)
except NoSuchElementException:
    source.append('-')
    
#destination:
try:
    rom = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[2]")        
    for i in rom:
        destinations=i.text
        destination.append(destinations)
except NoSuchElementException:
    destination.append('-')
    
#depature_time:
try:
    rom = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-4 top5 ']/span[1]")        
    for i in rom:
        dep_times=i.text
        dep_time.append(dep_times)
except NoSuchElementException:
    dep_time.append('-')
    
#arrival_time:
try:
    rom = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-3 top5']/span[1]")        
    for i in rom:
        arrival_times=i.text
        arrival_time.append(arrival_times)
except NoSuchElementException:
    arrival_time.append('-')
    
#duration:
try:
    rom = driver.find_elements_by_xpath("//span[@class='dura_md ng-binding']")        
    for i in rom:
        durations=i.text
        duration.append(durations)
except NoSuchElementException:
    duration.append('-')

#total_stops:
try:
    rom = driver.find_elements_by_xpath("//span[@class='dura_md2 ng-scope']")         
    for i in rom:
        total_stop=i.text.split(' Via')[0]
        total_stops.append(total_stop)
except NoSuchElementException:
    total_stops.append('-')

#flight_no:
try:
    rom = driver.find_elements_by_xpath("//span[@class='txt-r5']")        
    for i in rom:
        flight_nos=i.text
        flight_no.append(flight_nos)
except NoSuchElementException:
    flight_no.append('-')

# info is the block contatining all flight details
info = driver.find_elements_by_xpath("//div[@class='col-md-12 col-sm-12 main-bo-lis pad-top-bot ng-scope']")
print('Total length :',len(info))

# using loop to find the total values of flight details
for flight in range(len(info)):
    route_lst = []
    route=''
    journey.append('26/02/2022')         # journey date
    time.sleep(1)
    search_btn = driver.find_element_by_xpath("//div[@id='"+str(flight)+"']")
    search_btn.click()
    time.sleep(0.25)
    
    #prices of flight:
    try:
        info = driver.find_elements_by_xpath("//div[@class='col-md-2 col-sm-2 col-xs-5 mr5 cle']")
        pricy=info[flight].text.split('\n')[0]
        pp=pricy.replace(',','')
        price_info.append(pp)   
    except NoSuchElementException:
        price_info.append('-')
    info = driver.find_elements_by_xpath("//span[@class='txt-r9']")
    for leng in range(len(info)):
        route_lst.append(info[leng].text)
    for leng in range(len(route_lst)):
        if leng==0:
            route+=(route_lst[leng]) 
        else:
            route+=(route_lst[leng].split(route_lst[leng-1].split(' ')[2])[1])  #route of flight
    route_info.append(route)
    
```

```python
#1
da1=pd.DataFrame({'Airline': airline,'Journey Date': journey,'Source':source,'Destination': destination
                 ,'Route': route_info,'Dep_Time': dep_time,'Arrival_Time': arrival_time,'Duration': duration,
                 'Total_Stops': total_stops,'Flight_no': flight_no,'Price': price_info})
```
